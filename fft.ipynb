## Using the FFT to find the frequency of a reference signal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
frequency = 10
sampling_rate = 2**6
duration = 1
```

```python
t = np.linspace(0, duration, int(sampling_rate * duration))
```

```python
sawtooth_wave = signal.sawtooth(2 * np.pi * frequency * t)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(t, sawtooth_wave)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Wave Signal')
plt.grid(True)

plt.show()
```

```python
fft_result = np.fft.fft(sawtooth_wave)
n = len(fft_result)
fft_mag = np.abs(fft_result)
freq = np.fft.fftfreq(n, d = 1/sampling_rate)
```

```python
half_n = len(fft_mag) // 2
plt.figure(figsize=(12, 6))
plt.plot(fft_mag[:half_n])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT of the signal')
plt.grid(True)
```

```python
peak_index = np.argmax(fft_mag[:half_n])
peak_freq = freq[peak_index]
"Frequency: " + str(peak_freq) + " hz"
```
